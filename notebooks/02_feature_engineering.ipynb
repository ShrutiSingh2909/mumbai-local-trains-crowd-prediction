```python
import pandas as pd

# Load processed dataset
df = pd.read_csv("C:/Users/shruti/Documents/mumbai-local-trains-crowd-density-prediction/data/raw/mumbai_local_crowd_data.csv")

# Check data
print(df.shape)
df.head()
```

Create time-based features

```python
# Morning (6–11)
df["is_morning"] = df["hour"].apply(lambda x: 1 if 6 <= x <= 11 else 0)

# Evening (17–22)
df["is_evening"] = df["hour"].apply(lambda x: 1 if 17 <= x <= 22 else 0)

# Office hours (9–18)
df["is_office_hour"] = df["hour"].apply(lambda x: 1 if 9 <= x <= 18 else 0)

df[["hour", "is_morning", "is_evening", "is_office_hour"]].head(10)
```

Peak hour flag

```python
df["is_peak_hour"] = df["hour"].apply(
    lambda x: 1 if (8 <= x <= 10 or 18 <= x <= 21) else 0
)

df[["hour", "is_peak_hour"]].head(10)
```

Weekend Flag

```python
# Weekend flag
df["is_weekend"] = df["day_of_week"].apply(
    lambda x: 1 if x in ["Saturday", "Sunday"] else 0
)

df[["day_of_week", "is_weekend"]].head(10)
```

```python
# Encode Categorical Columns
from sklearn.preprocessing import LabelEncoder

# Create encoders
le_station = LabelEncoder()
le_day = LabelEncoder()
le_line = LabelEncoder()
le_crowd = LabelEncoder()

# Encode features
df["station_enc"] = le_station.fit_transform(df["station"])
df["day_enc"] = le_day.fit_transform(df["day_of_week"])
df["line_enc"] = le_line.fit_transform(df["line"])

# Encode target
df["crowd_enc"] = le_crowd.fit_transform(df["crowd_level"])

# Check encoding
df[["station", "station_enc", "day_of_week", "day_enc", "line", "line_enc", "crowd_level", "crowd_enc"]].head()
```

```python
# Save engineered dataset
df.to_csv("../data/processed/engineered_features.csv", index=False)

print("Engineered dataset saved successfully!")
```
